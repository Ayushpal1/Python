```python
import folium
import pandas
from geopy.geocoders import Nominatim
from geopy.geocoders import ArcGIS
```

```python
df = pandas.read_csv("MagicBricks.csv")
locations = list(df["Name"])
geocoder = ArcGIS()
l = []
for location in locations:
    newAddress = location[20:] + ", Pune, Maharashtra"
    l.append(geocoder.geocode(newAddress))
for item in l:
    print(item[1])
```

```python

map = folium.Map(location=[18.5204, 73.8567],zoom_start=12)
fg = folium.FeatureGroup(name = "test")
for item in l:
    marker  = folium.Marker(location=item[1])
    fg.add_child(marker)
map.add_child(fg)
map.save("map2.html")
map
```

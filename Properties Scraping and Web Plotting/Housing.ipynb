```python
import requests
from bs4 import BeautifulSoup
import pandas
```

```python
r = requests.get("https://housing.com/in/buy/searches/P3zgy19bgb2xjywqp")
c = r.content
```

```python
soup = BeautifulSoup(c,"html.parser")
```

```python
all = soup.find_all("article",{"class":"css-1nr7r9e"})
```

```python
l = []
for item in all:
    d = {}
    #to skip the featured article
    if item.find("div",{"class":"css-1ibkxbp"}):
        continue
    
    #to print the name and price of the property
    try:
        d["Name"]=item.find("h2",{"class":"css-1t3f68z"}).text
    except:
        d["Name"] = "Project Name not Found"
        
    try:
        d["Address"] = item.find("a",{"class":"css-16drx2b"}).text
    except:
        d["Address"] = "Address not found"
        
    d["Price"] = item.find("div",{"class":"css-18rodr0"}).text
    l.append(d)
df = pandas.DataFrame(l)
df
#df.to_csv("output.csv")
```
